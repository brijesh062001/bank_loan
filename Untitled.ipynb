```python
X=df.drop('y',axis=1)
y=df['y']
```

```python
plt.figure(figsize=(18,10))
sns.heatmap(df.corr(),cmap='viridis',annot=True)

plt.show()
```

```python
from sklearn.ensemble import ExtraTreesClassifier
ex=ExtraTreesClassifier()
ex.fit(X,y)
print(ex.feature_importances_)
```

```python
feat_importances = pd.Series(ex.feature_importances_, index=X.columns)
# considering top 5 important features
feat_importances.nlargest(5).plot(kind='barh')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(X_train,y_train)
```

```python
y_pred=log.predict(X_test)
```

```python
log.score(X_train,y_train)
```

```python
log.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,log_loss,accuracy_score
print('Confusion matrix\n',confusion_matrix(y_test,y_pred))
print('\nClassification report\n',classification_report(y_test,y_pred))
print('\nLog loss',log_loss(y_test,y_pred))
print('\nAccuracy score',accuracy_score(y_test,y_pred))
```
